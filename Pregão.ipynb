```python
# importanto biblioteca
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# declara o link
url = 'http://www2.bmf.com.br/pages/portal/bmfbovespa/lumis/lum-ajustes-do-pregao-ptBR.asp'
```

```python
# request código html
res = requests.get(url)
```

```python
html_doc = res.text
```

```python
# estrutura de dados aninhada
soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
```

```python
# pesquisa tag's
soup.table
```

```python
#extração e limpeza dos dados
dados = []
divs = soup.findAll("table", {"id": "tblDadosAjustes"})
for div in divs:
    rows = div.findAll('tr')
    for row in rows :
        dados.append(row.findAll('td'))
        
dados
```

```python
#cria um dataframe
df = pd.DataFrame(dados,columns=['Mercadoria',
                                 'Vencimento',
                                 'Preco_Anterior',
                                 'Preco_Atual',
                                 'Variacao',
                                 'Valor_ajuste_contrato'])
```

```python
df
```

```python
df.info()
```

```python
df.loc[:,'Mercadoria']= df['Mercadoria'].apply(lambda x: str(x).replace('<td>', ''))
df.loc[:,'Mercadoria']= df['Mercadoria'].apply(lambda x: str(x).replace('</td>', ''))
df.loc[:,'Mercadoria']= df['Mercadoria'].apply(lambda x: str(x).replace('<td class="text-center">', ''))

df.loc[:,'Vencimento']= df['Vencimento'].apply(lambda x: str(x).replace('<td>', ''))
df.loc[:,'Vencimento']= df['Vencimento'].apply(lambda x: str(x).replace('</td>', ''))
df.loc[:,'Vencimento']= df['Vencimento'].apply(lambda x: str(x).replace('<td class="text-center">', ''))

df.loc[:,'Preco_Anterior']= df['Preco_Anterior'].apply(lambda x: str(x).replace('<td>', ''))
df.loc[:,'Preco_Anterior']= df['Preco_Anterior'].apply(lambda x: str(x).replace('</td>', ''))
df.loc[:,'Preco_Anterior']= df['Preco_Anterior'].apply(lambda x: str(x).replace('<td class="text-right">', ''))

df.loc[:,'Preco_Atual']= df['Preco_Atual'].apply(lambda x: str(x).replace('<td>', ''))
df.loc[:,'Preco_Atual']= df['Preco_Atual'].apply(lambda x: str(x).replace('</td>', ''))
df.loc[:,'Preco_Atual']= df['Preco_Atual'].apply(lambda x: str(x).replace('<td class="text-right">', ''))

df.loc[:,'Variacao']= df['Variacao'].apply(lambda x: str(x).replace('<td>', ''))
df.loc[:,'Variacao']= df['Variacao'].apply(lambda x: str(x).replace('</td>', ''))
df.loc[:,'Variacao']= df['Variacao'].apply(lambda x: str(x).replace('<td class="text-right">', ''))

df.loc[:,'Valor_ajuste_contrato']= df['Valor_ajuste_contrato'].apply(lambda x: str(x).replace('<td>', ''))
df.loc[:,'Valor_ajuste_contrato']= df['Valor_ajuste_contrato'].apply(lambda x: str(x).replace('<td class="text-right">', ''))
df.loc[:,'Valor_ajuste_contrato']= df['Valor_ajuste_contrato'].apply(lambda x: str(x).replace('</td>', ''))
```

```python
df = df.replace(to_replace='None', value=np.nan).dropna()
df['Data_new'] = pd.to_datetime('now').date()
```
