```python
# importanto biblioteca
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# declara o link
url = 'https://www.cepea.esalq.usp.br/br/indicador/boi-gordo.aspx'
```

```python
# request código html
res = requests.get(url)
```

```python
html_doc = res.text
```

```python
# estrutura de dados aninhada
soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
```

```python
#extração e limpeza dos dados
dados = []
divs = soup.findAll("table", {"class": "imagenet-table imagenet-th2 imagenet-col-12"})
for div in divs:
    rows = div.findAll('tr')
    for row in rows :
        dados.append(row.findAll('td'))
        
dados
```

```python
#cria um dataframe
df = pd.DataFrame(dados,columns=['Data',
                                 'Valor_RS',
                                 'Var_Dia',
                                 'Var_Mês',
                                 'Valor_US'])
```

```python
print(df.head())
```

```python
df.info()
```

```python
df.loc[:,'Data']= df['Data'].apply(lambda x: str(x).replace('<td>', ''))
df.loc[:,'Data']= df['Data'].apply(lambda x: str(x).replace('</td>', ''))

df.loc[:,'Valor_RS']= df['Valor_RS'].apply(lambda x: str(x).replace('<td>', ''))
df.loc[:,'Valor_RS']= df['Valor_RS'].apply(lambda x: str(x).replace('</td>', ''))

df.loc[:,'Var_Dia']= df['Var_Dia'].apply(lambda x: str(x).replace('<td>', ''))
df.loc[:,'Var_Dia']= df['Var_Dia'].apply(lambda x: str(x).replace('</td>', ''))

df.loc[:,'Var_Mês']= df['Var_Mês'].apply(lambda x: str(x).replace('<td>', ''))
df.loc[:,'Var_Mês']= df['Var_Mês'].apply(lambda x: str(x).replace('</td>', ''))

df.loc[:,'Valor_US']= df['Valor_US'].apply(lambda x: str(x).replace('<td>', ''))
df.loc[:,'Valor_US']= df['Valor_US'].apply(lambda x: str(x).replace('</td>', ''))
```

```python
df = df.replace(to_replace='None', value=np.nan).dropna()
df['Data_new'] = pd.to_datetime('now').date()
df
```
