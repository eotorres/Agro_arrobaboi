```python
# importanto biblioteca
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# declara o link
url = 'https://www.scotconsultoria.com.br/cotacoes/boi-gordo/'
```

```python
# request código html
res = requests.get(url)
```

```python
html_doc = res.text
```

```python
# estrutura de dados aninhada
soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
```

```python
# pesquisa tag's
soup.table
```

```python
print(soup.find("tr",attrs={"conteudo"}).text)
```

```python
# extração e limpeza dos dados
dados = []
for x in soup.find_all("tr",attrs={"conteudo"}):
    dados.append(x.text.split('\n'))
```

```python
print(dados)
```

```python
#cria um dataframe
df = pd.DataFrame(dados,columns=['Data','Local','à vista (FUNRURAL)','C4','30 dias (FUNRURAL)','C6','#base²','à vista','C9','30 dias','C11','C12'])
```

```python
print(df.head())
```

```python
#exclui colunas em branco
df2 = df.drop(columns=['C4','C6','C9','C11','C12'])
```

```python
df2['Data'] = pd.to_datetime('now').date()
df2['Data_new'] = pd.to_datetime('now')
```

```python
from datetime import datetime
data_atual = datetime.today()
data_em_texto = data_atual.strftime('%d%m%Y')
data_em_texto
```
